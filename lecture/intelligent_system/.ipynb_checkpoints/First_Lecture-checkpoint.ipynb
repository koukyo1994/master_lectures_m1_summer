### 第一回 AIが社会に及ぼす影響について　イントロダクション

(素人の)議論はたくさんある。
AIに関する誤解に関する議論も少なくない。

ひとつの議論のの方向性として、人工知能の入り込んだ社会に関する議論がある。 
 一方で人工知能の技術そのものにかかる社会の要請という議論もあって然るべき。

ELSI: Ethical, Legal, and Social Issues ⇔ AI Technologies

人口知能を社会においてどう扱うかという議論は多い。 
 世の中の人が不安を持っているときに技術を持つものとしては何を言うべきかという議論はなされていない。

* 大学院の時間は短いので毎日2時間研究しかしない時間をとるといい
* メモを残しておくといい

### 人工知能に対して人々が抱く不安

* 人工知能が人類を滅ぼすのではないか?
* 人間の尊厳が脅かされるのではないか?
* 人工知能と人間の恋愛の可能性はどう考えればいいのか?

恋愛とは何かという議論になる。

* 人工知能が心を持つのか、持つとしたらそれをどう受け止めたらいいのか?

持たせたいのか? というのが答え。

* ロボットの権利や義務を考えることになるのか?

もう議論は始まっている。ロボットの作品に関する著作権など。法律はいかようにでも作れてしまうもの。

* 人の雇用を奪うのではないか?

大雑把な議論は多い。何をAIに任せたら嬉しくて何を任せては行けないのかはより正確な議論をする必要がある。

意外とゴミ収集のような部分は自動化が進んでいない。かえって高度に知的な仕事の方が置き換わる可能性が高い。
弁護士などはすでに置き換えられる可能性がある。(情報処理の部分)

日本は特に少子高齢化が進んでいるためよって場合によっては自動化してしまったほうが良い面もある。
単純に奪う/奪わないではなく、どのように役割分担するといいか、という議論が必要。

**チラシのように**論文を書く分野はAIに置き換わるかも。

* 人工知能の考えることと行うことを人が理解できなくなるのではないか

既に警察によって犯罪確率が高い/低いを推定すると言ったシステムも試験的に導入されている。

しかし、このようなシステムは現状ではブラックボックスになっている。
説明はできるようにすべきだし、出来無いにしてもするような努力はすべき。

* 人工知能の考えることと行うことを人が制御できなくなるのではないか?

電源を切ればいい、というのは無責任。
AIがあちこちに入り込んでいて複雑系をなしている可能性もある。ただ、電源を切るにしても様々な影響がありうる。

* 人工知能は想定外の事態に対処できないのではないか?

これは基本的には正しい。人間も基本的にはできない。
人工知能は学習した範囲でしか基本的には対応できない。

よくある解のひとつとして人間のオペレータと組み合わせるという可能性があるが、これは多くの場合には無意味であると考えられる。
普段から人工知能に任せきりで監視しているだけのオペレータが、人工知能でも対処できない事態に対処できるかどうかは怪しい。

むしろ普段から自分でオペレーションを行っている人のほうが想定外の事態には強いと考えられる。
AIが入手できないような状況に関する知識がある人間の専門家の優位性など。
マニュアルどおりに動くのではなく第一原理に立ち返って考えるなど、訓練を受けた人間はできるかも。

センサ情報を増やす、というのはひとつの解決策

* 軍事技術として応用されるとき、殺すことに対する心理的抵抗を減らしてしまうのではないか?

ヨーロッパはロボット兵器の開発は禁止すべきという立場、アメリカは全面的に禁止すべきという立場には反対。
アメリカの言い分としては、敵の軍事施設のみをピンポイントに攻撃することで、アメリカの兵士や敵国の民間人を殺さずにすむ。

アメリカの制服組は自動化に否定的。
命がかかっているからこそ戦争はできるだけ避けるのであって命がかからなくなったら歯止めが効かなくなる可能性。

* テロリストなどに悪用されるのではないか?

テロリストに使われるから、研究をやめるというのはナンセンス。
悪用を防ぐ研究をする、という立場。

* システムに侵入されて悪意を持った改変を施される恐れがあるのではないか?

できるだけシステムを透明にしておいて何が起こっているかわかりやすいようにすることで改変がすぐにわかるようにする、という対策。

* プライバシーが侵害されるのではないか? 

EU、日本、米国で立場が違う。
EUは厳格。日本は甘すぎるがそれを誤解している人も多い。

* 事故や失敗の責任を誰が取るのか? 

Uberの事故に関しても責任について話し合って欲しかった。

* 事故の賠償の保険制度をどのようにつくりかえることになるのか?

* 現行法規制と相容れない部分をどうするのか?

* どれくらい故障するのか、どのような失敗の恐れがあるのか?
